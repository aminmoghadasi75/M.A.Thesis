```python
def clear_list(l :str) :
    l =  l.replace("[", "").replace("]", "")
    l = l.replace("'", "")
    l = l.replace(" ", "")
    l = l.split(",")
    return l
```

```python
list_of_all_selected_features = []
folder_path = '/home/amin/thesis/src/Step/Feature_selection_Lasso_LogReg'
file_list = os.listdir(folder_path)
month = 0
for file in file_list:
    df = pd.DataFrame(pd.read_excel(f'{folder_path}/{file}'))
    df.drop(columns='Unnamed: 0', inplace=True)
    for item in df.iloc[-1] :
        month += 1
        p = clear_list(item)
        for i in p :
            list_of_all_selected_features.append(i)
```
