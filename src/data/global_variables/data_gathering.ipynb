```python
def clean_data(df):
    # Replace gaps forward from the previous valid value in: 'Oil Price', 'BTC Price' and 13 other columns
    df = df.fillna({'Oil Price': df['Oil Price'].ffill(), 'BTC Price': df['BTC Price'].ffill(), 'GAS Price': df['GAS Price'].ffill(), 'Cooper Price': df['Cooper Price'].ffill(), 'Aluminum Price': df['Aluminum Price'].ffill(), 'Gold Price': df['Gold Price'].ffill(), 'Treasury Bond': df['Treasury Bond'].ffill(), 'Treasury Yield 30 Years': df['Treasury Yield 30 Years'].ffill(), 'Treasury Yield 5 Years': df['Treasury Yield 5 Years'].ffill(), 'Ethereum-USD': df['Ethereum-USD'].ffill(), 'S&P 500': df['S&P 500'].ffill(), 'NASDAQ Composite': df['NASDAQ Composite'].ffill(), 'Dollar-Index': df['Dollar-Index'].ffill(), 'Lead Price': df['Lead Price'].ffill(), 'Steel Price': df['Steel Price'].ffill()})
    # Drop column: 'Ethereum-USD'
    df = df.drop(columns=['Ethereum-USD'])
    # Replace missing values with 0 in columns: 'Oil Price', 'BTC Price' and 11 other columns
    df = df.fillna({'Aluminum Price' : 0,'Oil Price': 0, 'BTC Price': 0, 'GAS Price': 0, 'Cooper Price': 0, 'Gold Price': 0, 'Treasury Bond': 0, 'Treasury Yield 30 Years': 0, 'Treasury Yield 5 Years': 0, 'S&P 500': 0, 'NASDAQ Composite': 0, 'Dollar-Index': 0, 'Lead Price': 0, 'Steel Price': 0})
    # Change column type to datetime64[ns] for column: 'Date'
    df = df.astype({'Date': 'datetime64[ns]'})
    # Change column type to float64 for columns: 'Oil Price', 'BTC Price' and 11 other columns
    df = df.astype({'Aluminum Price' : 'float64','Oil Price': 'float64', 'BTC Price': 'float64', 'GAS Price': 'float64', 'Cooper Price': 'float64', 'Gold Price': 'float64', 'Treasury Bond': 'float64', 'Treasury Yield 30 Years': 'float64', 'Treasury Yield 5 Years': 'float64', 'S&P 500': 'float64', 'NASDAQ Composite': 'float64', 'Dollar-Index': 'float64', 'Lead Price': 'float64', 'Steel Price': 'float64'})
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```
