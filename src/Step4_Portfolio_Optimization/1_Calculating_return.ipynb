### from portfolio --> finding the file --> finding the date --> finding the column'پایانی * سهم *'

```python
def get_next_month_values(df, date, column_name):
    # Convert the date to a pandas datetime object
    date = pd.to_datetime(date)

    # Calculate the first day of the next month
    first_day_next_month = date + pd.DateOffset(months=1) - pd.DateOffset(days=date.day - 1)

    # Calculate the last day of the next month
    last_day_next_month = first_day_next_month + pd.DateOffset(months=1) - pd.DateOffset(days=1)

    # Extract the values for the specified column and the next month
    next_month_values = df[(df.index >= first_day_next_month) & (df.index <= last_day_next_month)][column_name]

    # Return the first and last values of the next month
    return next_month_values.iloc[0], next_month_values.iloc[-1]
```

# RF - ANN

```python
final_result = {}
for date, value in portfolio.items() :
    date = date
    saving_return = []
    print(f'Working for {date} period')
    for stock in value :
        data_name = stock[0].replace('RfAnn_', '')
        print(f'{data_name} is processing ...')
        df = pd.read_excel(f'{data_folder}/{data_name}')
        df.set_index(df['Date'], inplace=True)
        first_value , last_value = get_next_month_values(df=df, date=date, column_name= 'پایانی*سهم*')
        return_value = (last_value - first_value) / first_value
        saving_return.append((data_name, return_value))
        print(f'The Return of {data_name} is {return_value} !')
    final_result[date] = saving_return
with open('./my_portfolio_return_RFANN.json', 'w') as f: # Save the portfolio
    json.dump(final_result, f)
```
